```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("experiments/results/compare_splitting_heuristics.json") as f:
    data = json.load(f)
    
    fig, ax = plt.subplots()
    ax.set_yscale('linear')
    ax.set_xscale('linear')
        
    for data in data:
        nb_spheres = np.array([experiment['nb_spheres'] for experiment in data['experiments']])
        results = [np.array(experiment['intersection_tests']) for experiment in data['experiments']]
        averages = [np.average(res) for res in results]
        stds = np.array([np.std(res) for res in results])

        
        ax.errorbar(nb_spheres, averages, yerr=stds, fmt='-o', label=data['splitting_heuristic'])
        
    ax.grid(True)
    ax.legend()
    plt.savefig('experiments/plots/compare_splitting_heuristics.png')
```
